```python
import pandas as pd
import numpy as np

# DataSets
from sklearn.datasets import fetch_openml
```

```python
X, y = fetch_openml('autos', version=1, as_frame=True, return_X_y=True)

# DataFrame
df = X
df['target'] = y

df.head()
```

```python
df=df[['price','num-of-doors','fuel-type']]

df.head()
```

```python
df['num-of-doors'].unique()
```

```python
df['fuel-type'].unique()
```

```python
# Pivot - round 2 decimal places
pivot = np.round(pd.pivot_table(df, values='price', index='num-of-doors', 
                                columns='fuel-type', aggfunc=np.mean), 2)
pivot
```
